```python
import os
import numpy as np
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
```

```python
data = pd.read_csv("/content/data.csv")
```

特徴量の選定

```python
audio_features = [
    'acousticness', 'danceability', 'energy', 'instrumentalness',
    'liveness', 'loudness', 'speechiness', 'tempo', 'valence'
]
```

特徴量の抽出と標準化

```python
X = data[audio_features]
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
X
```

KMeansクラスタリングを採用（20クラスタリング）

```python
kmeans = KMeans(n_clusters=20, random_state=42)
kmeans.fit(X_scaled)
```

クラスタラベルをつける

```python
data['cluster_label'] = kmeans.predict(X_scaled)
```

曲名検索を大文字小文字を無視して行えるようにする

```python
data['name_lower'] = data['name'].str.lower()
```

モデルの保存

```python
#各クラスの中心
np.save('kmeans_centroids.npy', kmeans.cluster_centers_)
#スケーラーのパラメータ
np.save('scaler_mean.npy', scaler.mean_)
np.save('scaler_scale.npy', scaler.scale_)
```

```python
data.to_csv("data_with_clusters.csv", index=False)
```
